```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras import regularizers
import random
%matplotlib inline
```

```python
data = tf.keras.datasets.mnist.load_data(path="mnist.npz")
```

```python
img = data[0][0][1]
pixels = np.asarray(img)
print(pixels.shape)
plt.imshow(pixels, cmap='gray')
```

```python
train_images = data[0][0]
train_labels = data[0][1]
test_images = data[1][0]
test_labels = data[1][1]
print(train_images.shape)
print(test_images.shape)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)), tf.keras.layers.Dense(128, activation='relu', kernel_regularizer = regularizers.l2(0.01)),tf.keras.layers.Dense(10)])
model.compile(optimizer="adam", loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs = 100)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
```

```python
#using l2 
print(test_acc)
```

```python
model2 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)), tf.keras.layers.Dense(128, activation='relu', ),tf.keras.layers.Dropout(0.25),tf.keras.layers.Dense(10)])
model2.compile(optimizer="adam", loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model2.fit(train_images, train_labels, epochs = 100)
```

```python
#using dropout
test_loss, test_acc = model2.evaluate(test_images, test_labels, verbose=0)
print(test_acc)
```

```python
def augment_images(train_images):
    for i in range(len(train_images)):
        chance = random.random()
        if chance<0.2:
            temp = train_images[i]
            np.transpose(temp)
            train_images[i]
    return train_images
    
```

```python
train_images = augment_images(train_images)
model3 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),tf.keras.layers.Dense(128, activation='relu'),tf.keras.layers.Dense(10)])
model3.compile(optimizer="adam", loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
model3.fit(train_images, train_labels, epochs = 100)
```

```python
#using data augmentation
test_loss, test_acc = model3.evaluate(test_images, test_labels, verbose=0)
print(test_acc)
```

```python
model4 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)), tf.keras.layers.Dense(128, activation='relu', kernel_regularizer = regularizers.l1(0.01)),tf.keras.layers.Dense(10)])
model4.compile(optimizer="adam", loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
model4.fit(train_images, train_labels, epochs = 100)
```

```python
#using l1 regression
test_loss, test_acc = model4.evaluate(test_images, test_labels, verbose=0)
print(test_acc)
```
